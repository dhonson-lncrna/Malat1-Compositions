```python
def base_counter(seq):
    '''Finds the base content of a sequence, assigns each base ratio
    to a variable, and returns a list with the contents'''
    for i in seq:
        ade = seq.count('A')/len(seq)
        thy = seq.count('T')/len(seq)
        cyt = seq.count('C')/len(seq)
        gua = seq.count('G')/len(seq)
    ls = [ade, thy, cyt, gua]
    return ls
        
```
